```python
import numpy as np 
```

## Load the model 

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model("digit_ann_model.keras")
```

# Building the canvas 

```python
import tkinter as tk
from PIL import Image , ImageDraw
root = tk.Tk()
root.title("Handwritten Digit Recognition")
root.geometry("400x400")
# CANVAS FOR DRAWING 
canvas = tk.Canvas(root, width = 300 , height = 300 , bg = "black")
canvas.pack(pady = 20)

# PIL image (same size as canvas )
image = Image.new("L" ,(300,300) , color =0)
draw_image = ImageDraw.Draw(image)


# function to draw on canvas
def draw(event): # this func runs everytime the mouse moves 
    x = event.x
    y = event.y
    r = 8 # brush radius
    canvas.create_oval(x-r , y-r , x+r , y+r , fill = "white" , outline = "white")
    # Draw on Pil IMage
    draw_image.ellipse((x-r, y-r, x+r, y+r) , fill = 255)
# bind mouse movt with left button pressed
canvas.bind("<B1-Motion>",draw)

# add a clear button to the canvas 
def clear_canvas():
    canvas.delete("all")
    draw_image.rectangle((0,0,300,300) , fill = 0)

#clear button
clear_btn = tk.Button(root , text = "clear" , command = clear_canvas)
clear_btn.pack(pady = 10)

result_label = tk.Label(root , text = "Draw a digit and click Predict", font=("Arial", 16))
result_label.pack(pady = 10)
def predict_digit():
    #1.  Resize the image: our canvas is (300 x 300)
    img = image.resize((28,28))

    # 2. convert the image to numpy array
    img_array = np.array(img)

    #3. Normalize (0,1)
    img_array = img_array / 255.0

    #4. Reshape for ANN
    img_array = img_array.reshape(1,28,28)

    # predict 
    prediction = model.predict(img_array)
    digit = np.argmax(prediction)
    
    print("Predicted digit:", digit)
    result_label.config(text=f"Predicted Digit: {digit}")


predict_btn = tk.Button(root, text="Predict", command=predict_digit)
predict_btn.pack(pady=5)

root.mainloop()
```
